Weight Logs:
   - Weight logs track users weight measurements over time. By analyzing weight log data, we can assess users progress towards weight management goals.


Objective:

 Weight log analysis helps users track their weight loss or gain journey and make informed decisions about diet and exercise.

1. Importing all needed libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Importing Dataset

```python
df = pd.read_csv("/content/weightLogInfo_merged.csv")
```

3. Analysing Dataset

Checking consistency of Dataset

```python
print(df.isna().sum())
# Fill nan values with 0.
df['Fat'] = df['Fat'].fillna(0)

# Check for duplicate values
print(df.duplicated().sum())
```
