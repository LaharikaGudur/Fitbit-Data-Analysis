 Heart rate data:

  Provides insights into users cardiovascular health and exercise intensity levels. By monitoring heart rate during physical activity, we can determine the intensity of workouts and assess users cardiovascular fitness.

Objective:

 Analyzing heart rate trends helps tailor exercise routines to optimize cardiovascular health and fitness outcomes.


1. import required libraries.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Read the given data .

```python
df = pd.read_csv("/content/heartrate_seconds_merged.csv")
```

3. Check information about the data.

```python

print(df.shape)
df.info()
```

```python
# Display first 10 rows of the data.
df.head(10)
```

```python
# Convert the time column into datetime format.
df['Time'] = pd.to_datetime(df['Time'])
```

```python
# Check the nan values.
print(df.isna().sum())

# Drop the nan value.
df = df.dropna(axis=0)
```

```python
# Download the final cleaned data.
heartrate_seconds_cleaned_data = df
heartrate_seconds_cleaned_data.to_csv('heartrate_seconds_cleaned_data.csv',index=False)
from google.colab import files
files.download('heartrate_seconds_cleaned_data.csv')
```

**Exploratory Data Analysis**

```python
# Distribution plot of Heartrate values.
sns.distplot(df['Value'],color='g')
plt.title('Distribution of Heart Rate Values')
plt.show()
```

```python
# Box plot of 'Value' by hour
df['Hour'] = df['Time'].dt.hour
plt.figure(figsize=(10, 6))
sns.boxplot(x='Hour', y='Value', data=df)
plt.title('Box Plot of Value by Hour')
plt.xlabel('Hour')
plt.ylabel('Value')
plt.show()
```
