Daily Activity:
   - The daily activity data provides insights into users overall physical activity levels. It includes metrics such as total steps taken, calories burned, and exercise intensity levels. By analyzing daily activity trends, we can identify patterns and variations in users' activity levels over time.

Objectives:

 By analyzing daily activity trends, we can identify patterns and variations in users activity levels over time.

1. Import all required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Loading Dataset

```python
# Reading the given data.
daily_data = pd.read_csv("/content/dailyActivity_merged.csv")
```

3.Analysis of dataset

```python
# Displaying first 10 rows of the data.
daily_data.head(10)
```

4. Checking Dataset

```python
# Check the shape of data.
print(daily_data.shape)

# Checking for the Nan values.
print(daily_data.isna().sum())

# Checking datatype of each column in data.
print(daily_data.dtypes)

# Converting the activitydate into datetime data type.
daily_data['ActivityDate'] = pd.to_datetime(daily_data['ActivityDate'])

# Checking for the duplicate values.
daily_data.duplicated().sum()
```

**Summary of dailyActivity_merged data**
1.   Checked the shape of the dataset . (940,15)
2.   Checked whether the null values are present or not. There are no
     null/nan values in the dataset.
3.   Checked the data type of each column. Converted the datatype of
     ActivityDate into datatime datatype format.
4.   Checked for the duplicate values. No duplicate values are present.

 **Exploratory Data Analysis**



```python
sns.heatmap(daily_data.corr(),cbar=True,linewidths=0.5)
plt.show()
```

Key Finding:

*   There is positive correlation between Calories and TotalActive Minutes,Total Steps and Total Distance.
*   There is positive correlation between Calories and SedentaryMinutes.



```python
sns.distplot(daily_data.Calories,bins=10,color='g')
plt.show()
```

```python
sns.scatterplot(x='TotalDistance',y='TotalSteps',data=daily_data,color='red')
plt.title('Total Distance vs Total Steps')
plt.show()
```

```python
sns.histplot(daily_data['TotalSteps'])
plt.title('Distribution of Total Steps')
plt.xlabel('Total Steps')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='TotalSteps',y='Calories',data=daily_data,color='violet')
plt.title('Total Steps vs Calories')
plt.show()
```

```python
fig,axes = plt.subplots(2,2)
sns.scatterplot(x='VeryActiveMinutes',y='Calories',data=daily_data,ax=axes[0,0])
sns.scatterplot(x='FairlyActiveMinutes',y='Calories',data=daily_data,ax=axes[0,1])
sns.scatterplot(x='LightlyActiveMinutes',y='Calories',data=daily_data,ax=axes[1,0])
sns.scatterplot(x='SedentaryMinutes',y='Calories',data=daily_data,ax=axes[1,1])
plt.tight_layout()
plt.show()
```

Key Finding:

*   By this exploratory data analysis we can find the active minutes and inactive(that is sedentary, Lightly Active, Fairly active)minutes.
